```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/apple_sales_2024.csv')
data.head()
```

```python
print(data.columns)
```

```python
summary = data.describe()
print("Statistical Summary: ")
print(data.describe())
```

```python
print(data.describe(include=['object']))
```

```python
plt.hist(data['State'], bins=10, edgecolor='black', color='skyblue')
plt.title('Distribution of State')
plt.xlabel('Frequency')
plt.ylabel('State')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['Region'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Distribution of Region')
plt.xlabel('Region')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['iPhone Sales (in million units)'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot of iPhone Sales (in million units)')
plt.ylabel('iPhone Sales (in million units)')
plt.show()
plt.show()
```

```python
sns.pairplot(data[['State', 'Region', 'iPhone Sales (in million units)', 'iPad Sales (in million units)']].dropna())
plt.show()
```

```python
correlation_matrix = data[['iPhone Sales (in million units)', 'iPad Sales (in million units)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Region', data=data, palette='pastel')
plt.title('Distribution of Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Region', y='iPhone Sales (in million units)', data=data, palette='pastel')
plt.title('Distribution of iPhone Sales by Region')
plt.show()
```
