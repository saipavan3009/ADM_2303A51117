```python
import pandas as pd
import sklearn as sk
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data=pd.read_csv('/content/survey lung cancer.csv')
data.head()
```

```python
print("missing values before imputation:")
print(data.isnull().sum())
```

```python
data['SMOKING'].fillna(data['SMOKING'].mean(), inplace=True)
```

```python
data['AGE'].fillna (data['AGE'].mean(), inplace=True)
data['SMOKING'].fillna(data['SMOKING'].mode()[0], inplace=True)
```

```python
print("\nMissing values after imputation:")
print(data.isnull().sum())
```

```python
scaler_minmax = MinMaxScaler()
columns_to_normalize = ['AGE', 'SMOKING']
data[columns_to_normalize]= scaler_minmax.fit_transform(data[columns_to_normalize])
```

```python
data.head()
```

```python
scaler_standard = StandardScaler()
data[columns_to_normalize] = scaler_standard.fit_transform(data[columns_to_normalize])
```
