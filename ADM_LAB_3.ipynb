```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/iris")

print("Path to dataset files:", path)
```

```python
iris = load_iris()
x=iris.data
y=iris.target
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y,cmap='viridis',edgecolors='k',s=50)
plt.title('PCA of Iris Dataset (2 Components)')
plt.xlabel('First Principal Component 1')
plt.ylabel('Second Principal Component 2')
plt.colorbar(label='target Class')
plt.show()
```

```python
print("Explained Variance Ratio of Each Component:")
print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=3)
x_pca = pca.fit_transform(x_scaled)
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111, projection='3d')
ax.scatter(x_pca[:,0], x_pca[:,1], x_pca[:, 2], c=y, cmap='viridis', edgecolor='k', s=100)
ax.set_xlabel('First Principal Component 1')
ax.set_ylabel('Second Principal Component 2')
ax.set_zlabel('Third Principal Component 3')
plt.show()
```
